```python
import pandas as pd
import seaborn as sns
```

```python
readsinfo_path = '/global/scratch/projects/fc_wolflab/bbioinfo/plate1_info.csv'
colslist = ['FASTQ Path - Read 1',  'Total Reads', 'Best Genome Match', 'Strain_ID', 'MP16SAbbr', 'Library Name', 'Avg Q Score Read 1','Avg Q Score Read 2']
readsinfo_df = pd.read_csv(readsinfo_path, thousands=',', usecols=colslist)
readsinfo_df['qscores_averaged'] = (readsinfo_df['Avg Q Score Read 1'] + readsinfo_df['Avg Q Score Read 2'])/2
readsinfo_df.head()
```

```python
readsinfo_df['Total Reads'].head()
```

```python
sns.histplot(data=readsinfo_df, x='Total Reads', log_scale=True)
```

```python
sns.histplot(data=readsinfo_df, x='qscores_averaged', log_scale=True)
```

```python
sns.histplot(data=readsinfo_df, x='Avg Q Score Read 1', log_scale=True)
```

```python
readsinfo_df.describe()
```
